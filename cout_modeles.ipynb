```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "n-gram": {
        "params": [1.94, 85.4],
        "loss": [2.39, 1.81]
    },
    "MLP": {
        "params": [4.48, 8.54, 16.7, 32.9, 65.4, 353, 1360],
        "loss": [1.96, 1.84, 1.75, 1.69, 1.61, 1.38, 1.12]
    },
    "RNN": {
        "params": [7.78, 27.8, 105, 406, 1600],
        "loss": [1.88, 1.72, 1.61, 1.47, 1.22]
    },
    "Transformer": {
        "params": [19.5, 71.2, 274, 1070, 4200],
        "loss": [1.76, 1.65, 1.39, 1.03, 0.92]
    },
    "Mamba": {
        "params": [11.5, 35.7, 123, 450, 1720],
        "loss": [1.81, 1.70, 1.59, 1.44, 1.29]
    }
}
```

```python
plt.figure(figsize=(6, 4))

for model_family, values in data.items():
    plt.plot(values["params"], values["loss"], marker='o', label=model_family)

plt.xlabel("Nombre de paramètres (milliers)")
plt.ylabel("Coût")
plt.title("Coût des différentres architectures \n en fonction du nombre de paramètres")
plt.legend()

plt.xscale('log')
```

```python
plt.figure(figsize=(8, 6))

for model_family, values in data.items():
    sns.lineplot(x=values["params"], y=values["loss"], marker='o', label=model_family)

plt.xlabel("Nombre de paramètres (milliers)")
plt.ylabel("Coût")
plt.title("Coût des différentes architectures \n en fonction du nombre de paramètres")

plt.legend(title='Architectures')

plt.xscale('log')
plt.tight_layout()
plt.savefig("comp_architectures.png", dpi=600)
```
