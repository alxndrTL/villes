```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import square
```

```python
# Paramètres
f = 10
T = 1 / f
fs = 5000  # Fréquence d'échantillonnage
t = np.linspace(0, 30 * T, int(3 * T * fs), endpoint=False)

signal_principal = square(2 * np.pi * f * t)
signal_A = square(2 * np.pi * (f/10) * t)
signal_B = square(2 * np.pi * (f/10) * t + np.pi/2)
```

```python
# Tracé des signaux
plt.figure(figsize=(12, 8))
plt.subplot(311)
plt.title('Signal Carré Principal à fréquence f')
plt.plot(t, signal_principal)
plt.ylim(-1.5, 1.5)

plt.subplot(312)
plt.title('Signal Carré A à fréquence f/10')
plt.plot(t, signal_A)
plt.ylim(-1.5, 1.5)

plt.subplot(313)
plt.title('Signal Carré B à fréquence f/10 déphasé de 90°')
plt.plot(t, signal_B)
plt.ylim(-1.5, 1.5)

plt.tight_layout()
plt.show()
```
